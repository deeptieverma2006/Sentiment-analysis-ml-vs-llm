```python

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

train_df = pd.read_csv("train_clean.csv")
test_df = pd.read_csv("test_clean.csv")

y_train = train_df["label"]
y_test = test_df["label"]

print("Train size:", train_df.shape)
print("Test size:", test_df.shape)
print(train_df["filtered_text"][0][:200])
```

```python
#Bag of words
import time
cv=CountVectorizer(max_features=5000)
start=time.time()
X_train_cv=cv.fit_transform(train_df["filtered_text"])
X_test_cv=cv.transform(test_df["filtered_text"])
end=time.time()
print("Count Vectorization done")
print("Train matrix shape: ",X_train_cv.shape)
print("Test matrix shape: ",X_test_cv.shape)
print(f"Time taken: {end-start:.2f} seconds")
```

```python
print(X_train_cv[:5, :10].toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=5000)
start=time.time();
X_train_tfidf=tfidf.fit_transform(train_df["filtered_text"])
X_test_tfidf=tfidf.transform(test_df["filtered_text"])
end=time.time();
print("Train matrix shape",X_train_tfidf.shape)
print("Test matrix shape", X_test_tfidf.shape)
print(f"Time taken: {end-start:.2f} seconds")
```

```python
import gensim.downloader as api
glove_model=api.load("glove-wiki-gigaword-100")
print("Done")          
```

```python
def review_to_vector(review,model,vector_size=100):
    words=review.split()
    vectors=[model[word] for word in words if word in model]
    if len(vectors)==0:
        return np.zeros(vector_size)
    return np.mean(vectors, axis=0)
start=time.time()
X_train_glove=np.array([review_to_vector(review,glove_model) for review in train_df["filtered_text"]])
X_test_glove=np.array([review_to_vector(review,glove_model) for review in test_df["filtered_text"]])
end=time.time();
print("Train matrix shape: ",X_train_glove.shape)
print("Test matrix shape: ",X_test_glove.shape)
print(f"Time taken: {end-start:.2f} seconds")
```

```python
np.save("X_train_glove.npy", X_train_glove)
np.save("X_test_glove.npy", X_test_glove)
```

```python
from scipy import sparse
sparse.save_npz("X_train_cv.npz", X_train_cv)
sparse.save_npz("X_test_cv.npz", X_test_cv)
```

```python
sparse.save_npz("X_train_tfidf.npz", X_train_tfidf)
sparse.save_npz("X_test_tfidf.npz", X_test_tfidf)
```
